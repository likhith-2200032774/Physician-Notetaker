```python
import spacy
import re
import json
from transformers import pipeline
nlp = spacy.load("en_core_web_sm")
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
sentiment_model = pipeline("sentiment-analysis")
transcript = """
Physician: Good morning, Ms. Jones. How are you feeling today?

Patient: Good morning, doctor. I’m doing better, but I still have some discomfort now and then.

Physician: I understand you were in a car accident last September. Can you walk me through what happened?

Patient: Yes, it was on September 1st, around 12:30 in the afternoon. I was driving from Cheadle Hulme to Manchester when I had to stop in traffic. Out of nowhere, another car hit me from behind, which pushed my car into the one in front.

Physician: That sounds like a strong impact. Were you wearing your seatbelt?

Patient: Yes, I always do.

Physician: What did you feel immediately after the accident?

Patient: At first, I was just shocked. But then I realized I had hit my head on the steering wheel, and I could feel pain in my neck and back almost right away.

Physician: Did you seek medical attention at that time?

Patient: Yes, I went to Moss Bank Accident and Emergency. They checked me over and said it was a whiplash injury, but they didn’t do any X-rays. They just gave me some advice and sent me home.

Physician: How did things progress after that?

Patient: The first four weeks were rough. My neck and back pain were really bad—I had trouble sleeping and had to take painkillers regularly. It started improving after that, but I had to go through ten sessions of physiotherapy to help with the stiffness and discomfort.

Physician: That makes sense. Are you still experiencing pain now?

Patient: It’s not constant, but I do get occasional backaches. It’s nothing like before, though.

Physician: That’s good to hear. Have you noticed any other effects, like anxiety while driving or difficulty concentrating?

Patient: No, nothing like that. I don’t feel nervous driving, and I haven’t had any emotional issues from the accident.

Physician: And how has this impacted your daily life? Work, hobbies, anything like that?

Patient: I had to take a week off work, but after that, I was back to my usual routine. It hasn’t really stopped me from doing anything.

Physician: Everything looks good. Your neck and back have a full range of movement, and there’s no tenderness or signs of lasting damage.
Physician: Given your progress, I’d expect you to make a full recovery within six months of the accident.
"""

def extract_medical_entities(text):
    doc = nlp(text)
    entities = {"SYMPTOMS": [], "DIAGNOSIS": [], "TREATMENT": [], "PROGNOSIS": []}
    if "whiplash" in text.lower():
        entities["DIAGNOSIS"].append("Whiplash injury")
    if "pain" in text.lower():
        entities["SYMPTOMS"].extend(["Neck pain", "Back pain"])
    if "physiotherapy" in text.lower():
        entities["TREATMENT"].append("Physiotherapy sessions")
    if "painkiller" in text.lower():
        entities["TREATMENT"].append("Painkillers")
    if "full recovery" in text.lower():
        entities["PROGNOSIS"].append("Full recovery expected within six months")
    for ent in doc.ents:
        if ent.label_ in ["ORG", "DATE", "TIME", "GPE"]:
            entities.setdefault("OTHER_INFO", []).append(f"{ent.text} ({ent.label_})")
    for k in entities:
        entities[k] = list(set(entities[k]))
    return entities

def extract_keywords(text, top_n=10):
    words = re.findall(r'\b[a-zA-Z]{4,}\b', text.lower())
    stopwords = spacy.lang.en.stop_words.STOP_WORDS
    words = [w for w in words if w not in stopwords]
    freq = {}
    for w in words:
        freq[w] = freq.get(w, 0) + 1
    sorted_words = sorted(freq.items(), key=lambda x: -x[1])
    return [w for w, _ in sorted_words[:top_n]]

medical_entities = extract_medical_entities(transcript)
keywords = extract_keywords(transcript)

summary = summarizer(transcript, max_length=150, min_length=40, do_sample=False)[0]["summary_text"]

def analyze_sentiment_intent(text):
    patient_lines = [line for line in text.splitlines() if line.strip().startswith("Patient:")]
    joined_text = " ".join(patient_lines)
    sentiment_res = sentiment_model(joined_text)[0]["label"]
    intent = []
    lower = joined_text.lower()
    if any(x in lower for x in ["pain", "hurt", "ache"]):
        intent.append("Reporting symptoms")
    if any(x in lower for x in ["worried", "concerned", "hope", "afraid"]):
        intent.append("Seeking reassurance")
    if any(x in lower for x in ["thank", "relief", "better", "improving"]):
        intent.append("Expressing gratitude / reassurance")
    if not intent:
        intent.append("General update")
    return {"Sentiment": sentiment_res, "Intent": intent}

sentiment_output = analyze_sentiment_intent(transcript)

def build_soap(text, entities):
    soap = {
        "Subjective": {
            "Chief_Complaint": ", ".join(entities.get("SYMPTOMS", [])) or "Not specified",
            "History_of_Present_Illness": "Patient was in a car accident, experienced neck and back pain for four weeks, improved after physiotherapy."
        },
        "Objective": {
            "Physical_Exam": "Full range of motion, no tenderness.",
            "Observations": "Patient occasionally experiences mild back pain."
        },
        "Assessment": {
            "Diagnosis": ", ".join(entities.get("DIAGNOSIS", [])) or "Whiplash injury suspected",
            "Severity": "Mild, improving"
        },
        "Plan": {
            "Treatment": entities.get("TREATMENT", []),
            "Follow_Up": "Return if pain worsens or persists beyond six months."
        }
    }
    return soap

soap_note = build_soap(transcript, medical_entities)

structured_summary = {
    "Patient_Name": "Janet Jones",
    "Symptoms": medical_entities["SYMPTOMS"],
    "Diagnosis": medical_entities["DIAGNOSIS"],
    "Treatment": medical_entities["TREATMENT"],
    "Prognosis": medical_entities["PROGNOSIS"],
    "Keywords": keywords,
    "Summary": summary,
    "Sentiment_Intent": sentiment_output
}

print("=== 🧾 Structured Summary ===")
print(json.dumps(structured_summary, indent=2))
print("\n=== 🩺 SOAP Note ===")
print(json.dumps(soap_note, indent=2))
```
